```python
import sys
sys.path.append('/Users/markrice/Documents/GitHub/RGH_NLP/')


from DataTools import network_datatools as nw, ps_datatools as ps, DB_datatools as db
import pandas as pd
import numpy as np
import itertools
import timeit
```

```python
[curs, conn] = db.db_connect('DataTools/drugbank.db')
```

```python
entity_list = ps.ps_read_entitylist('DataTools/CFS_entities.csv')
alias_list = nw.findaliases(entity_list)
```

```python
start_time = timeit.default_timer()
# code you want to evaluate
answers = []
for i in alias_list:
    answers.append(db.db_findrelations(curs, i))
elapsed = timeit.default_timer() - start_time
print("Time taken = ", elapsed)
```

```python
answers = [item for item in answers if item]
```

```python
hash = {k:v for k, v in zip(list(entity_list['Name']), alias_list)}
```

```python
answers_list = []
for row in answers:
    
    for subrow in row:
        target_aliases = db.db_findtargetsynonyms(curs,subrow[2])
        
        lst_key= []
        for item in target_aliases:
            x = [key for key, value in hash.items() if item[0].casefold() in value.casefold()]
            if x:
                lst_key.append(x)
        lst_key = list(itertools.chain.from_iterable(lst_key))
        if lst_key:
            lst_key = lst_key[0]
            subrow[2] = lst_key
            answers_list.append(subrow)
                
        print(subrow)
        print(target_aliases)
        print(lst_key)
        print("\n")
```

```python
drugedges = pd.DataFrame(answers_list, columns=['Source', 'Interaction', 'Target'])
```

```python
drugedges.to_csv("CFS_Drug_Edges.csv", index=False)
```
